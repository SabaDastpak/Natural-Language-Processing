```python
robert_frost='robert_frost.txt'
```

```python
# cleaningg the data
import string
lines=[]
with open(robert_frost,"r", encoding="utf-8") as f:
    for line in f:

        line=line.rstrip().lower()
        if line:
            line=line.translate(str.maketrans('','',string.punctuation))
            lines.append(line)
```

```python

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt', quiet=True)
```

```python
# tokenize data and add START and END to each line
list_of_tokens=[]
for line in lines:
    tokens=word_tokenize(line)
    tokens.insert(0,'START')
    tokens.insert(0,'START')
    tokens.append("END")
    list_of_tokens.append(tokens)
```

```python
list_of_tokens[0:3]
```

```python
#  creatiing the probability dictionary --- first order Markov Model

probs = {}

for tokens in list_of_tokens:
    for i in range(len(tokens) - 2):
        word_one = tokens[i]
        word_two = tokens[i + 1]
        word_three=tokens[i+2]

        if (word_one, word_two) not in probs:
            probs[(word_one, word_two)] = {}

        if word_three not in probs[(word_one, word_two)]:
            probs[(word_one, word_two)][word_three] = 1
        else:
            probs[(word_one, word_two)][word_three] += 1
```

```python
probs[('START','START')].keys()
```

```python
#  generating the poem

import random
def generate_line(probs):
    current_state=('START','START')
    result=[]

    while True:
        next_words=list(probs[current_state].keys())
        counts=list(probs[current_state].values())
        next_word=random.choices(next_words, weights=counts, k=1)[0]

        if next_word=="END":
            break

        result.append(next_word)
        current_state=(current_state[1],next_word)

    return " ".join(result)
```

```python
# to generate multiple lines:
for _ in range(4):
    print(generate_line(probs))
```
