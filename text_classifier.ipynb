```python
def file_to_list(txt_file):
    lines=[]
    with open(txt_file,'r',encoding="utf-8") as f:
        for line in f:
            line=line.strip()
            if line:
                lines.append(line)
    return lines
```

```python
texts=[]
labels=[]

# 0 represents robert frost , 1 represents edgar allan poe

frost= file_to_list(robert_frost)


for i in frost:
    texts.append(i)
    labels.append(0)

poe=file_to_list(edgar_allan_poe)

for i in poe:
    texts.append(i)
    labels.append(1)
```

```python
x_train_frost = [x_train_S[i] for i in range(len(y_train)) if y_train[i] == 0]
x_train_poe   = [x_train_S[i] for i in range(len(y_train)) if y_train[i] == 1]
```

```python
#  computing pi for frost
N=len(x_train_frost)
V=len(word2idx)
pi_frost=np.zeros(V)

for i in x_train_frost:
    if len(i)>0:
        h=i[0]
        pi_frost[h]+=1
pi_frost/=N

#  computing pi for poe

M=len(x_train_poe) 
pi_poe=np.zeros(V)

for i in x_train_poe:
    if len(i)> 0:
        L=i[0]
        pi_poe[L]+=1
pi_poe/=M

```

```python
# computing A
#  computing add_one smoothed A for frost
A_frost_smoothed=np.zeros((V,V))
H_frost_smoothed=np.zeros(V)
for seq in x_train_frost:
    for i in range(len(seq)-1):
        A_frost_smoothed[seq[i], seq[i+1]] += 1
        H_frost_smoothed[seq[i]]+=1

```

```python
######## add_one smoothing for pi
#  computing add_one smoothined pi for frost


N_frost=len(x_train_frost)
V=len(word2idx)
pi_frost_smoothed=np.zeros(V)

for i in x_train_frost:
    if len(i)>0:
        h=i[0]
        pi_frost_smoothed[h]+=1
pi_frost_smoothed+=1
pi_frost_smoothed/=(N_frost+V)

#  computing add_one smoothed pi for poe

N_poe=len(x_train_poe) 
pi_poe_smoothed=np.zeros(V)

for i in x_train_poe:
    if len(i)> 0:
        L=i[0]
        pi_poe_smoothed[L]+=1
pi_poe_smoothed+=1
pi_poe_smoothed/=(N_poe+V)


```

```python
#  sanity check
pi_frost_smoothed.sum()   # ≈ 1.0
pi_poe_smoothed.sum()     # ≈ 1.0
```

```python
A_frost_smoothed+=1
H_frost_smoothed+=V

for i in range(V):
    if H_frost_smoothed[i] > 0:
        A_frost_smoothed[i] /= H_frost_smoothed[i]

# computing add_one smoothed A for poe

A_poe_smoothed=np.zeros((V,V))
H_poe_smoothed=np.zeros(V)
for seq in x_train_poe:
    for i in range(len(seq)-1):
        A_poe_smoothed[seq[i], seq[i+1]] += 1
        H_poe_smoothed[seq[i]]+=1

A_poe_smoothed+=1
H_poe_smoothed+=V

for i in range(V):
    if H_poe_smoothed[i] > 0:
        A_poe_smoothed[i] /= H_poe_smoothed[i]
```

```python
# p(author=frost)
probability_author_frost= len(x_train_frost)/(len(x_train_frost)+len(x_train_poe))

# p(author=poe)
probability_author_poe= len(x_train_poe)/(len(x_train_frost)+len(x_train_poe))
poem=input("enter the poem:")

def string_to_seq(poem):
    poem_seq=[]
    tokens = word_tokenize(poem)
    for token in tokens:
        idx = word2idx.get(token, 0)
        poem_seq.append(idx)
    return poem_seq

poem_seq=string_to_seq(poem)

# p(poem|frost)
probability_frost=np.log(pi_frost_smoothed[poem_seq[0]])
for i in range(len(poem_seq)-1):

    probability_frost+=np.log(A_frost_smoothed[poem_seq[i],poem_seq[i+1]])


# p(poem|poe)
probability_poe=np.log(pi_poe_smoothed[poem_seq[0]])
for i in range(len(poem_seq)-1):

    probability_poe+=np.log(A_poe_smoothed[poem_seq[i],poem_seq[i+1]])


probability_frost += np.log(probability_author_frost)
probability_poe   += np.log(probability_author_poe)


# Final decision (argmax)
if probability_frost > probability_poe:
    print("Author: Frost")
else:
    print("Author: Poe")
```
